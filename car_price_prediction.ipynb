```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load data
df = pd.read_csv('car data.csv')
df.head()
```

```python
#drop car_name column
df=df.drop(columns=['Car_Name'])
df.head()
```

```python
#convert categorical to numerical
df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.head()
```

```python
x=df_encoded.drop('Selling_Price',axis=1)
y=df_encoded['Selling_Price']
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#train ml model randomforest
from sklearn.ensemble import RandomForestRegressor
model =RandomForestRegressor(
    n_estimators=200,
    criterion='squared_error',
    random_state=42
)
model.fit(x_train,y_train)
```

```python
#evaluate model mse mae
from sklearn.metrics import mean_absolute_error
y_pred=model.predict(x_test)
mae=mean_absolute_error(y_test,y_pred)
print("mae",mae)
from sklearn.metrics import mean_squared_error
y_pred=model.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
print("mse",mse)
from sklearn.metrics import r2_score
y_pred=model.predict(x_test)
print("r2")
r2_score(y_test,y_pred)
```

```python
all_model_features = x_train.columns.tolist()

def predict_car_price(
    year: int,
    present_price: float,
    kms_driven: int,
    fuel_type: str,
    owner: int,
    company: str,
    seller_type: str,
    transmission: str
) -> float:
  input_data_df = pd.DataFrame(0, index=[0], columns=all_model_features)
  input_data_df['Year'] = year
  input_data_df['Present_Price'] = present_price
  input_data_df['Kms_Driven'] = kms_driven
  input_data_df['Owner'] = owner
  fuel_type_col = f'Fuel_Type_{fuel_type}'
  if fuel_type_col in all_model_features:
      input_data_df[fuel_type_col] = 1
  company_col = f'company_{company}'
  if company_col in all_model_features:
      input_data_df[company_col] = 1
  if seller_type == 'Individual':
      if 'Seller_Type_Individual' in all_model_features:
          input_data_df['Seller_Type_Individual'] = 1
  if transmission == 'Manual':
      if 'Transmission_Manual' in all_model_features:
          input_data_df['Transmission_Manual'] = 1
  prediction = model.predict(input_data_df)
  return prediction[0]

sample_year = 2014
sample_present_price = 5.59
sample_kms_driven = 27000
sample_fuel_type = 'Petrol'
sample_owner = 0
sample_company = 'maruti suzuki'
sample_seller_type = 'Dealer'
sample_transmission = 'Manual'

predicted_selling_price = predict_car_price(
    sample_year,
    sample_present_price,
    sample_kms_driven,
    sample_fuel_type,
    sample_owner,
    sample_company,
    sample_seller_type,
    sample_transmission
)

print(f"Predicted Selling Price: {predicted_selling_price:.2f} Lakhs")
```
